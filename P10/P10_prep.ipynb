```python
c1 = list(df1['Country'])
c2 = list(df2['CountryName'])
caps = list(df2['CapitalName'])
m = {}
for i in range(len(c2)):
    m[c2[i]] = caps[i]
```

```python
len(caps)
```

```python
caps1 = []
for country in c1:
    if country in m:
        caps1.append(m[country])
    else:
        print(country)
        caps1.append("Unknown")
```
